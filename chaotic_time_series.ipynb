```python
import numpy as np
import matplotlib.pyplot as plt

def logistic_map(r, x):
    return r * x * (1 - x)

def generate_chaos(r, x0, num_iterations):
    chaos = [x0]
    for _ in range(num_iterations):
        x = logistic_map(r, chaos[-1])
        chaos.append(x)
    return chaos

# Parameters
r = 3.9    # Control parameter
x0 = 0.1   # Initial value
num_iterations = 1000

# Generate chaotic time series
chaos = generate_chaos(r, x0, num_iterations)

# Plot the chaotic time series
plt.figure(figsize=(10, 6))
plt.plot(chaos, 'b-', lw=0.5)
plt.title('Chaotic Time Series (Logistic Map)')
plt.xlabel('Iteration')
plt.ylabel('Value')
plt.show()
```

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

def lorenz(t, xyz, sigma, rho, beta):
    x, y, z = xyz
    dxdt = sigma * (y - x)
    dydt = x * (rho - z) - y
    dzdt = x * y - beta * z
    return [dxdt, dydt, dzdt]

# Parameters
sigma = 10.0
rho = 28.0
beta = 8.0 / 3.0

# Initial conditions
xyz0 = [1.0, 1.0, 1.0]

# Time span
t_span = (0, 100)
num_points = 10000
t_eval = np.linspace(t_span[0], t_span[1], num_points)

# Solve the differential equations
sol = solve_ivp(lorenz, t_span, xyz0, method='LSODA', t_eval=t_eval, args=(sigma, rho, beta))

# Extract the solution
x, y, z = sol.y

# Plot the Lorenz attractor
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, lw=0.5)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Lorenz Attractor')
plt.show()
```

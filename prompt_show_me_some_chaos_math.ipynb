```python
# prompt: show me some chaos math

# Here's the classic logistic map
def logistic_map(x, r):
  return r * x * (1 - x)

# Here's how to visualize the results of the logistic map
def plot_logistic_map(r, x0, num_iterations):
  x = x0
  for i in range(num_iterations):
    plt.plot(x, 'o')
    x = logistic_map(x, r)
  plt.show()

# Let's plot the results for a few different values of `r`
plot_logistic_map(3.57, 0.5, 1000)
plot_logistic_map(3.59, 0.5, 1000)
plot_logistic_map(3.6, 0.5, 1000)
```
